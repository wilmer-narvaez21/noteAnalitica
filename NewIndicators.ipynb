#### 3. Número de intentos medio por asignatura

```python
#Número de intentos medio por asignatura
DataInt = NewData[(NewData["Matricula"] == 'E14081394') & (NewData["Programa de estudios"] == "ISIC-2010-224-INGENIERÍA EN SISTEMAS COMPUTACIONALES")]
#DataInt = DataInt.assign(Num_Intentos = DataInt.groupby("Nombre de la Materia").size())

NumIntentos = DataInt[["Nombre de la Materia"]].value_counts().to_frame("Num. Intentos")#Datos repetinos y contiene recuentos de las filas
#DataInt.groupby('Nombre de la Materia').size().to_frame()
#NumIntentos
#.loc() indexación basada en etiquetas

NumIntentos["Curso"] = ''
NumIntentos.loc[NumIntentos["Num. Intentos"] == 1, "Curso"] = "Normal"
NumIntentos.loc[NumIntentos["Num. Intentos"] == 2, "Curso"] = "Repeticion"
NumIntentos.loc[NumIntentos["Num. Intentos"] == 3, "Curso"] = "Especial"
NumIntentos
```

#### 6. Tasa de rendimiento (créditos aprobados / matriculados). Se puede especificar por tiempo
semestre consta de 230 creditos en total
(Matriculados)total de creditos que se a escrito una vez terminado el semestre,
sacar indicador por semestre, aprobados / matriculados
ejemplo llevo 30 creditos y aprobo 15

```python
Rendimiento = NewData[(NewData["Matricula"] == 'E14081394') & (NewData["Programa de estudios"] == "ISIC-2010-224-INGENIERÍA EN SISTEMAS COMPUTACIONALES") &
                      (NewData["Ciclo escolar de la materia"]%2 != 0)]
Rendimiento
```

```python
#Sacar el promedio de cada semestre
```

```python
#nombres de la columnas con las que quiero ver el resultado, en este caso con creditos
#no incluir verano
creditosSuma = Rendimiento.groupby(['Semestre'])[['Creditos']].sum()
creditosSuma
```
